# Qiskit Aer Simulators

## Noise free Simulator

```python
import numpy as np
# Import Qiskit
import qiskit as qk
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram, plot_state_city
import qiskit.quantum_info as qi
```

```python
simulator = AerSimulator()
```

```python
print(AerSimulator.available_methods(simulator))
print(AerSimulator.available_devices(simulator))
```

```python
circ = QuantumCircuit(2)
circ.h(0)
circ.cx(0, 1)
circ.measure_active() # measure all (active) qubits

# Transpile for simulator
simulator = AerSimulator()
circ = transpile(circ, simulator)
display(circ.draw(output='mpl'))
```

### Counting the outcomes of the measurements

Using default inputs of $|00\rangle$

Available methods are `.run(circuit, shots, memory)` to run a circuit. 

Then you can use `.result()` to get the results of the run. The results can be accessed in a few ways:
- `.get_counts(circuit)` to get the counts of the measurements
- `.get_memory(circuit)` to get the memory of the circuit

```python
# Run and get counts
sim_result = simulator.run(circ, shots = 1000, memory=True).result()
sim_counts = sim_result.get_counts(circ)
plot_histogram(sim_counts, title='Bell-State counts')
```

### Returning measurement outcomes for each count

```python
memory = sim_result.get_memory(circ)
print(memory)
```
