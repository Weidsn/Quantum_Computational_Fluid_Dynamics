```python
import qiskit as qk
from qiskit import QuantumCircuit
import numpy as np

print(f"Qiskit version: {qk.__version__}")
```

```python
qc_0 = qk.QuantumCircuit(2)
qc_0.h(0)
qc_0.cx(0, 1)

display(qc_0.draw('mpl'))
```

```python
# Create a quantum circuit with 4 qubits
theta = np.pi / 4  # Example angle for rotation

qc = QuantumCircuit(4)

# Add a Hadamard gate on qubit 0
qc.rx(theta, 0)  # Add a rotation around the X-axis on qubit 0
qc.ry(theta, 1)  # Add a rotation around the Y-axis on qubit 1
qc.rz(theta, 2)  # Add a rotation around the Z-axis on qubit 2
qc.h(3)  # Add a Hadamard gate on qubit 3

qc.x(1)  # Add an X gate on qubit 3
qc.y(2)  # Add a Y gate on qubit 2
qc.z(3)  # Add a Z gate on qubit 1

# Add a CNOT gate on qubits 0 and 1
qc.cx(0, 1)
qc.cy(1, 2)  # Add a CNOT gate on qubits 0 and 2
qc.cz(2, 3)  # Add a CNOT gate on qubits 0 and 3
qc.ccx(0, 1, 2)  # Toffoli gate (CCX) on qubits 0, 1, and 2

qc.measure_all()  # Measure all qubits

# Draw the circuit
display(qc.draw('mpl'))
```
